```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a DataFrame
data = pd.read_csv(r"C:\Data Manipulation Technique/Top2000CompaniesGlobally.csv")

# Top 10 companies by Sales, Profits, and Market Value
top_sales = data.nlargest(10, 'Sales ($billion)')
top_profits = data.nlargest(10, 'Profits ($billion)')
top_market_value = data.nlargest(10, 'Market Value ($billion)')

# Setting up the visualization style
sns.set_style("whitegrid")

# Visualizing the top 10 companies
fig, ax = plt.subplots(3, 1, figsize=(14, 18))

# Top 10 Companies by Sales
sales_plot = sns.barplot(data=top_sales, x='Sales ($billion)', y='Company', ax=ax[0], palette='Blues_d')
ax[0].set_title('Top 10 Companies by Sales ($billion)')
ax[0].set_xlabel('Sales ($billion)')
ax[0].set_ylabel('Company')

for p in sales_plot.patches:
    sales_plot.annotate(f"{p.get_width():.2f}", (p.get_width(), p.get_y() + p.get_height() / 2),
                        ha='center', va='center', xytext=(3, 0), textcoords='offset points')

# Top 10 Companies by Profits
profits_plot = sns.barplot(data=top_profits, x='Profits ($billion)', y='Company', ax=ax[1], palette='Greens_d')
ax[1].set_title('Top 10 Companies by Profits ($billion)')
ax[1].set_xlabel('Profits ($billion)')
ax[1].set_ylabel('Company')

for p in profits_plot.patches:
    profits_plot.annotate(f"{p.get_width():.2f}", (p.get_width(), p.get_y() + p.get_height() / 2),
                          ha='center', va='center', xytext=(3, 0), textcoords='offset points')

# Top 10 Companies by Market Value
market_value_plot = sns.barplot(data=top_market_value, x='Market Value ($billion)', y='Company', ax=ax[2], palette='Purples_d')
ax[2].set_title('Top 10 Companies by Market Value ($billion)')
ax[2].set_xlabel('Market Value ($billion)')
ax[2].set_ylabel('Company')

for p in market_value_plot.patches:
    market_value_plot.annotate(f"{p.get_width():.2f}", (p.get_width(), p.get_y() + p.get_height() / 2),
                               ha='center', va='center', xytext=(3, 0), textcoords='offset points')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a DataFrame
data = pd.read_csv(r"C:\Data Manipulation Technique/Top2000CompaniesGlobally.csv")

# Count of companies by Continent
continent_counts = data['Continent'].value_counts()

# Top 10 countries with the most companies
top_countries = data['Country'].value_counts().head(10)

# Setting up the visualization style
sns.set_style("whitegrid")

# Visualizing the distribution
fig, ax = plt.subplots(2, 1, figsize=(14, 14))

# Distribution by Continent
continent_plot = sns.barplot(x=continent_counts.index, y=continent_counts.values, ax=ax[0], palette='viridis')
ax[0].set_title('Number of Companies by Continent')
ax[0].set_xlabel('Continent')
ax[0].set_ylabel('Number of Companies')

for p in continent_plot.patches:
    continent_plot.annotate(f"{int(p.get_height())}", (p.get_x() + p.get_width() / 2, p.get_height()),
                            ha='center', va='bottom', xytext=(0, 5), textcoords='offset points')

# Top 10 Countries by Number of Companies
countries_plot = sns.barplot(x=top_countries.index, y=top_countries.values, ax=ax[1], palette='inferno')
ax[1].set_title('Top 10 Countries by Number of Companies')
ax[1].set_xlabel('Country')
ax[1].set_ylabel('Number of Companies')
ax[1].tick_params(axis='x', rotation=45)

for p in countries_plot.patches:
    countries_plot.annotate(f"{int(p.get_height())}", (p.get_x() + p.get_width() / 2, p.get_height()),
                            ha='center', va='bottom', xytext=(0, 5), textcoords='offset points')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
#4 Correlation Analysis

# Correlation Analysis
correlation_matrix = data[['Sales ($billion)', 'Profits ($billion)', 'Assets ($billion)', 'Market Value ($billion)']].corr()

# Visualizing the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, square=True)
plt.title('Correlation Matrix')
plt.show()
```
